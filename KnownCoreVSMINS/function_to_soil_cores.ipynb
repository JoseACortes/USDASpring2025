```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import soilcoresim as scs
```

```python
import json
with open('data/field_info.json') as f:
    field_info = json.load(f)
```

```python
positions = pd.read_csv('data/positions.csv')
```

```python
positions
```

```python
x = positions['x']
y = positions['y']
```

```python
fence = np.array(field_info['corners'])
plt.figure(figsize=(10, 6))
plt.plot(fence[:, 0], fence[:, 1], 'k-', label='fence')
plt.plot(x, y, 'ro', label='soil core')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Field boundary')
plt.axis('equal')
plt.show()
```

```python
# pick random points inside the fence

n = 15000
x = np.random.uniform(fence[:, 0].min(), fence[:, 0].max(), n)
y = np.random.uniform(fence[:, 1].min(), fence[:, 1].max(), n)

plt.figure(figsize=(10, 6))
plt.plot(fence[:, 0], fence[:, 1], 'k-', label='fence')
plt.plot(x, y, 'ro', label='soil core')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Field boundary')
plt.axis('equal')
plt.show()
```
